```python
def get_cicle_data(n=50, r=1):
    
    np.random.seed(n)
    
    thetas = np.random.uniform(0, 2*np.pi, n)
    data = np.array([r * np.sin(thetas), r * np.cos(thetas)]).T + np.random.normal(scale=1/8, size=(n,2))
    
    return data
```

```python
def get_gaussian_data(n, mean=0.0, sigma=1.0):
    
    np.random.seed(n)
    
    data = np.random.normal(loc=mean, scale=sigma, size=(n,2))
    
    return data
```

```python
def PCA(xs, l):
    
    x_average = xs.mean(0)
    #x_average = (np.sum(data[:,0])/ len(data), np.sum(data[:,1])/ len(data))
    # compute x_tilde = x_i - x_average
    x_tilde1 = []
    x_tilde2 = []
    for i in range (len(xs)):
        x_tilde1.append(xs[i,0] - x_average[0])
        x_tilde2.append(xs[i,1] - x_average[1])
    
    X = np.vstack((x_tilde1, x_tilde2)).T

    #plt.scatter(x_tilde[:,0], x_tilde[:,1], c=labels)
    #covariance matrix
    C = np.dot(X.T, X)
    #print(C.shape)
    #eigendecomposition -> ordered descending
    eigenvalues, eigenvectors = scipy.linalg.eigh(C)
    eigenvalues_desc = np.flip(eigenvalues, axis=0)
    #l largest eigenvectors
    V_l = eigenvectors[-l:]
    y_i = np.dot(V_l, X.T)
    P = np.dot(V_l.T, V_l)
    #print(np.dot(P, X.T).shape)
    z_i = np.dot(P, X.T).T + x_average
    
    return y_i, z_i, eigenvalues_desc, V_l
```

```python
y_i, z_i, eigenvalues_desc, V_l = PCA(data, 1)
plt.figure()
plt.scatter(z_i[:,0], z_i[:,1], c=labels)
```

### b)

```python
def kernel_PCA(xs, kernel, l=2):
    n = len(kernel)
    one_n = [[1.0/n for i in range(n)] for j in range(n)]
    K_tilde = kernel - np.dot(one_n, kernel) - np.dot(kernel, one_n) + np.dot(np.dot(one_n, kernel), one_n)
    C = np.dot(K_tilde.T, K_tilde)
    eigenvalues, eigenvectors = scipy.linalg.eigh(C)
    eigenvalues, eigenvectors = eigenvalues.real, eigenvectors.real
    V_l = eigenvectors/np.sqrt(abs(eigenvalues))
    Y = []
    ys = 0
    
    for i in list(range(n)):
        for j in list(range(l)):
            ys = ys + np.dot(V_l[i, j], K_tilde[j, i])
        Y.append(ys)
        ys = 0
    return Y
```

Taking $$ u = \begin{pmatrix} -1 \\ -1 \end{pmatrix}, v = \begin{pmatrix} 1 \\ 1 \end{pmatrix}\in \mathbb{R}^2, W = \begin {pmatrix} -1 & 1\end{pmatrix} \in \mathbb{R}^{1x2} $$
$$Wu = 0$$
$$Wv = 0$$
$$Wu = Wv$$
$$ u \neq v$$

An exact recovery is not possible because u and v are mapped to the same value. Therefore multiplying any of these values with U will also lead to the same value, eliminating the previously existing difference between u and v making exact recovery impossible.

```python
def MDS(D, d):

    # Number of points  
    # d: num of dimensions
    n = len(D)

     
    # matrix S
    S = np.zeros((n,n))
    
    for i in range(n):
        for j in range(n):
            S[i][j] = 0.5*(D[1][i]**2 + D[1][j]**2 - D[i][j]**2)
    
   
    #eigendecomposition
    eigenvalues, eigenvectors = eigh(S)
    
    # largest d 
    V_d = eigenvectors[:,-d:] 
    Val_d = eigenvalues[-d:]
    diag = np.diag(Val_d)
    X = np.dot(V_d, np.sqrt(diag))
 
    return X
```
